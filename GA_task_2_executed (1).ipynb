```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
from IPython.display import display

# Load Stable Diffusion v1.5
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Prompt (change this to create different images)
prompt = "A futuristic cityscape at sunset with flying cars, digital art style"

# Generate image
image = pipe(prompt).images[0]

# Save and display
image.save("my_generated_image.png")
display(image)   # <-- This will show image in Colab output
```
